```python
import pandas as pd

# List of CSV files
csv_files = ['2019.csv','2018.csv','2017.csv', '2016.csv', '2015.csv']

# Empty list to store cleaned dataframes
cleaned_dfs = []

# Loop through each CSV file
for file in csv_files:
    # Load the dataset
    df = pd.read_csv(file)
    
    # Perform data cleaning steps
    
    # Save the cleaned dataframe to a new CSV file
    cleaned_file_name = f"cleaned_{file}"
    df.to_csv(cleaned_file_name, index=False)
    
    # Append the cleaned dataframe to the list
    cleaned_dfs.append(df)
```

```python
import pandas as pd

# List of CSV files
csv_files = ['2019.csv','2018.csv','2017.csv', '2016.csv', '2015.csv']

# Empty list to store dataframes
dfs = []

# Load each dataset and display variable information
for file in csv_files:
    # Load the dataset
    df = pd.read_csv(file)
    
    # Display variable information
    print(f"Variables in {file}:\n")
    print(df.info())
    print("\n")

    # Display basic descriptive statistics
    print(f"Basic Descriptive Statistics for {file}:\n")
    print(df.describe())
    print("\n")
    
    # Append the dataframe to the list
    dfs.append(df)
```
